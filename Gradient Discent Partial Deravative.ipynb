```python
import seaborn as sns

import matplotlib.pyplot as plt 

import numpy as np

import pandas as pd
```

```python
data = pd.read_csv("regression.csv")
```

```python
#第一題

X = data.iloc[:,0].values

y = data.iloc[:,1].values

data.corr()
```

```python
#2 第二題


def computeCost(X,y):
    h = temp[0] + X*temp[1]
    return (0.5) * np.sum((h-y)**2)
```

```python


# 第三題 3-1. gradient descent algorithm 
# 3-2. 畫出散點圖(1.Seaborn , 2.matplotlib)
# 3-3. Cost Function , Iterations 加分題

theta = [0,0] 
eta = 0.01   # learning Rate
color = 'green' 

count = 200

x_axis = np.arange(count)




cost = []   
temp = [0,0]
for i in range(1,count+1,1): 
    
    
    
    h = temp[0] + X*temp[1]
    
    temp[0] = temp[0] - eta * (h - y).sum()
    temp[1] = temp[1] - eta * np.sum(((h - y)* X))
    error = computeCost(X,y)
    cost.append(error)


   
```

```python
#3-2使用Seaborn版本 :

plt.figure(figsize=(15,7))
sns.regplot(X,y,scatter=True,fit_reg=True)

plt.xlabel('Variable X',fontsize=18)
plt.ylabel('Variable y',fontsize=18)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.title('Variable X  &  y  Correlation',fontsize=18)


plt.show()
```

```python
#3-2 使用matplotlib 版本 :

plt.figure(figsize=(15,7))
plt.plot(X , y , 'o')
plt.ylim([200,700])
plt.plot(X ,temp[0] + temp[1]*X)
plt.xlabel('Variable X',fontsize=18)
plt.ylabel('Variable y',fontsize=18)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.title('Variable X  &  y  Correlation',fontsize=18)
plt.show()
```

```python
# 3-3. Cost Function , Iterations 加分題

plt.figure(figsize=(14,7))
plt.plot(x_axis , cost , color=color , label='ETA = 0.01')

plt.xlabel('Iteration Times',fontsize=18)
plt.ylabel('Cost Function',fontsize=18)
plt.title('Gradient Discent & Iterations',fontsize=18)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)


plt.legend(fontsize='x-large')

plt.show()
```
